```python
import numpy as np
import matplotlib.pyplot as plt

# Ein Vektor ist nur eine Liste von Zahlen
v = np.array([2, 1])

# Plotten (Visualisierung ist alles!)
plt.figure()
plt.arrow(0, 0, v[0], v[1], head_width=0.2, head_length=0.2, color='red')
#plt.arrow(0, 0, v[0], v[1])
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.grid()
plt.show()
```

```python
# Eine 2x3 Matrix (2 Zeilen, 3 Spalten)
A = np.array([[1, 2, 3], 
              [4, 5, 6]])

# Ein Vektor mit 3 Elementen
x = np.array([1, 0, 1])

# Das Dot-Product (Skalarprodukt)
ergebnis = np.dot(A, x) 
print(f"Shape von A: {A.shape}")
print(f"Shape von x: {x.shape}")
print(f"Ergebnis: {ergebnis}")
# Überlege: Warum kommt ein Vektor mit 2 Elementen heraus?
```

```python
import numpy as np

# A: Unsere Daten (Der Input)
# Wir simulieren 100 Datensätze (Zeilen). 
# Jeder Datensatz hat 3 Eigenschaften (Spalten).
# Shape: (100, 3) -> (Batch Size, Features)
Data = np.random.rand(100, 3) 

# B: Unser Modell (Die Gewichte/Transformation)
# Es transformiert 3 Eingangs-Eigenschaften in 2 Ausgangs-Werte.
# Shape: (3, 2) -> (Input Features, Output Features)
Weights = np.random.rand(3, 2)

# C: Die Berechnung (Matrix Multiplikation)
# np.dot macht hier die "Zeile-mal-Spalte" Rechnung für alle 100 Zeilen auf einmal!
Predictions = np.dot(Data, Weights)

print(f"Shape der Daten:       {Data.shape}")
print(f"Shape der Gewichte:    {Weights.shape}")
print(f"Shape der Vorhersage:  {Predictions.shape}")

# Schau dir mal die erste Vorhersage an (erste Zeile)
print(f"\nVorhersage für Datensatz 1: {Predictions[0]}")
```
